```python
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math 
pd.set_option('display.max_columns', 500)

df = pd.read_csv('CaseStudyData.csv')
df.Variable5 = df.Variable5.map({'blue':1, 'green':0})
df.Variable8 = df.Variable8.map({'square':0, 'round':1, 'triangle':2})
df = df[df.Variable4 != ' -   ']
df.Variable4 = df.Variable4.astype('float')
df = df.sample(frac=1)
```

```python
fc = [
      'Variable1',
      'Variable2',
      'Variable4',
      'Variable6']

CF = df[fc]
```

```python
CF = CF.astype(float)

for x in CF.columns:
    for i in range(6):
        var1 = x + '**3'
        CF[var1.format(i)] = pow(np.maximum(CF[x],0),3)
        var2 = x + '**2'
        CF[var2.format(i)] = pow(np.maximum(CF[x],0),2)
        var3 = x + '**1/3'
        CF[var3.format(i)] = pow(np.maximum(CF[x],0),1/3)
        var4  = x + "**1/2"
        CF[var4.format(i)] = pow(np.maximum(CF[x],0),1/2)
        var5 = x + 'log(x+1)'
        CF[var5.format(i)] = np.log10(np.maximum(CF[x],0)+1)
        var6 = x + 'max(x,0)'
        CF[var6.format(i)] = np.maximum(CF[x],0)
```

```python
CF['Variable5'] = df.Variable5
CF['Variable8'] = df.Variable8
CF['Variable9'] = df.Variable9
CF['Variable3'] = df.Variable3
CF['Variable7'] = df.Variable7
```

```python
features = CF.values
target = df.TargetVariable.values
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.linear_model import LogisticRegression as LG
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
import matplotlib.pyplot as plt

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.30, random_state=143,stratify=target)
clf = LG(C=100, solver='lbfgs')
lgsf = SFS(estimator=clf, k_features=9, forward=True, scoring='roc_auc', cv=None, n_jobs=6, floating=True,verbose=1)
lgsf.fit(X_train, y_train)

print('best combination (ROC: %.3f): %s\n' % (lgsf.k_score_, lgsf.k_feature_idx_))
print('all subsets:\n', lgsf.subsets_)
plot_sfs(lgsf.get_metric_dict())
```

```python
#CF['TargetVariable'] = df.TargetVariable
data = pd.DataFrame(CF.columns)
var = list(data.iloc[list(lgsf.k_feature_idx_),0])
var
```

```python
#var = [#'Variable2',
       #'Variable6',
       #'Variable2**1/2',
       #'Variable2log(x+1)',
#       'Variable4log(x+1)',
       #'Variable6**2',
#       'Variable5',
 #'Variable3',
#       'Variable7']

var = [#'Variable6',
       'Variable1**3',
       'Variable1**2',
       'Variable2log(x+1)',
 #'Variable4**1/2',
       'Variable5',
 #'Variable8',
       'Variable3',
       'Variable7']

X_train1 = pd.DataFrame(X_train, columns=CF.columns)
X_test1 = pd.DataFrame(X_test, columns=CF.columns)
X_train1 = X_train1[var]
X_test1 = X_test1[var]
```

```python
clf1 = LG(solver='lbfgs')
clf1.fit(X_train1,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = clf1.predict(X_test1)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()

mat = confusion_matrix(y_test, y_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
from sklearn import metrics
print(metrics.classification_report(y_pred, y_test))
```

```python
y_prob = clf1.predict_proba(X_test1)[:,1]
y_probT = clf1.predict_proba(X_train1)[:,1]
```

```python
from matplotlib.backends.backend_pdf import PdfPages

X_train1['status'] = np.transpose(y_train)
X_train1['pred'] = np.transpose(y_probT)

X_test1['status'] = np.transpose(y_test)
X_test1['pred'] = np.transpose(y_prob)


with PdfPages('TeleLR.pdf') as pdf:
    for x in X_train1.columns:
        if x != 'test':
        
            X_train1['test'] = pd.cut(X_train1[x].rank(method='average'),10, labels=False)
            X_train2 = X_train1.groupby(['test'], as_index=False).mean()
            X_test1['test'] = pd.cut(X_test1[x].rank(method='average'),10, labels=False)
            X_test2 = X_test1.groupby(['test'], as_index=False).mean()
            fig = plt.figure(x)
            
            if x != 'status' and x != 'pred':
                
                plt.scatter(X_train2[x], X_train2.status, label = 'Train')
                plt.plot(X_train2[x], X_train2['pred'], label = 'Train_Prediction')
                plt.scatter(X_test2[x], X_test2.status, label = 'Test')
                plt.plot(X_test2[x], X_test2['pred'], label = 'Test_Prediction')
                plt.xlabel(x)
                plt.ylabel('P(app)')
                plt.legend()
                plt.title('P(app) as function of '+str(x))
                
            if x == 'pred':
        
                plt.plot([0, 1], [0, 1], "k:")
                plt.plot(X_train2[x], X_train2.status, 'o-', label = 'Train')
                
                plt.plot(X_test2[x], X_test2.status,'o-', label = 'Test')
                plt.xlabel('Predicted P(app)')
                plt.ylabel('Actual P(app)')
                #plt.xlim(0.2, 0.8)
                #plt.ylim(0.2, 0.8)
                plt.legend()
                plt.title('Callibration Curve for P(app)')
        
            if x == 'status':
               
                plt.hist(y_probT, range=[0, 1], bins=20, histtype="step", lw=2, label = 'Train')
                plt.hist(y_prob, range=[0, 1], bins=20, histtype="step", lw=2, label = 'Test')
                #plt.xlim(0.2, 0.8)
                plt.ylabel('Count')
                plt.xlabel('P(app)')
                plt.legend()
                plt.title('Histogram of P(app)')
            
            pdf.savefig(fig)
```

```python
from sklearn import metrics

fpr, tpr, _ = metrics.roc_curve(y_test, y_prob)
auc = metrics.auc(fpr,tpr)

with PdfPages('ROC.pdf') as pdf2:
    fig = plt.figure()
    title = 'ROC with AUC =' + auc.astype('str')
    plt.plot(fpr, tpr)
    plt.plot([0,1],[0,1],'g-')
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title(title)
    plt.show()
    pdf2.savefig(fig)

print('AUC = ', auc)
```
