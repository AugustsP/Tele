```python
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math 
pd.set_option('display.max_columns', 500)

df = pd.read_csv('CaseStudyData.csv')
df.Variable5 = df.Variable5.map({'blue':1, 'green':0})
df.Variable8 = df.Variable8.map({'square':0, 'round':1, 'triangle':2})
df = df[df.Variable4 != ' -   ']
df.Variable4 = df.Variable4.astype('float')
#df = df.drop('Variable4', axis=1)
#df = df.sample(frac=1)
```

```python
df['FN'] = 80000
df['FP'] = 1000
df['TP'] = 0
df['TN'] = 0
```

```python
#dg = df.head(481)
#tests = df.tail(119)

dg = df.head(472)
tests = df.tail(119)
```

```python
dgcostm = dg[['FN','FP','TP','TN']]
testscostm = tests[['FN','FP','TP','TN']]
dg = dg.drop(['FN','FP','TP','TN'],axis=1)
tests = tests.drop(['FN','FP','TP','TN'],axis=1)

dgcostm = dgcostm.values
testscostm = testscostm.values
```

```python
features = dg[['Variable4','Variable3','Variable7','Variable5','Variable9']].values
#features = dg[['Variable3','Variable5','Variable7']].values
#features = dg.loc[:,'Variable1':'Variable9'].values
target =  dg['TargetVariable'].values
```

```python
from sklearn import cross_validation
from sklearn.ensemble import GradientBoostingClassifier as GB
from sklearn.metrics import log_loss
from sklearn import cross_validation

skf = cross_validation.StratifiedShuffleSplit(target, n_iter=5)
results = []
resultsT = []

for train, test in skf:

    X_train, X_test = features[train], features[test]
    y_train, y_test = target[train], target[test]

    clf = GB(learning_rate=0.05, n_estimators=68)    
    clf.fit(X_train, y_train)
    prob = clf.predict_proba(X_test)[:,1]
    probT = clf.predict_proba(X_train)[:,1]
    cv_score_mli = log_loss(y_test, prob)
    cv_score_mliT = log_loss(y_train, probT)
    results.append(cv_score_mli)
    resultsT.append(cv_score_mliT)
    print(cv_score_mli)
    #print(cv_score_mliT)
```

```python
print(np.array(results).mean(),' +- ',np.array(results).std()/np.sqrt(10))
print(np.array(resultsT).mean(),' +- ',np.array(resultsT).std()/np.sqrt(10))
```

```python
featuresT = tests[['Variable4','Variable3','Variable7','Variable5','Variable9']]
#featuresT = tests[['Variable3','Variable5','Variable7']]
#featuresT = tests.loc[:,'Variable1':'Variable9']
targetT =  tests['TargetVariable'].values
```

```python
from sklearn.metrics import accuracy_score
from costcla import savings_score
from costcla import models

y_pred = clf.predict(featuresT)
print(accuracy_score(targetT, y_pred))
print(savings_score(targetT, y_pred, testscostm))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()

mat = confusion_matrix(targetT, y_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
y_prob = clf.predict_proba(featuresT)[:,1]
y_probT = clf.predict_proba(features)[:,1]
```

```python
from matplotlib.backends.backend_pdf import PdfPages

features = pd.DataFrame(features, columns=featuresT.columns)
features['status'] = np.transpose(target)
features['pred'] = np.transpose(y_probT)

featuresT['status'] = np.transpose(targetT)
featuresT['pred'] = np.transpose(y_prob)

with PdfPages('TeleGBM1.pdf') as pdf:
    for x in features.columns:
        if x != 'test':
        
            features['test'] = pd.cut(features[x].rank(method='average'),10, labels=False)
            features1 = features.groupby(['test'], as_index=False).mean()
            featuresT['test'] = pd.cut(featuresT[x].rank(method='average'),10, labels=False)
            featuresT1 = featuresT.groupby(['test'], as_index=False).mean()
            fig = plt.figure(x)
            
            if x != 'status' and x != 'pred':
                
                plt.scatter(features1[x],features1.status, label = 'Train')
                plt.plot(features1[x], features1['pred'], label = 'Train_Prediction')
                plt.scatter(featuresT1[x], featuresT1.status, label = 'Test')
                plt.plot(featuresT1[x], featuresT1['pred'], label = 'Test_Prediction')
                plt.xlabel(x)
                plt.ylabel('P(app)')
                plt.legend()
                plt.title('P(app) as function of '+str(x))
                
            if x == 'pred':
        
                plt.plot([0, 1], [0, 1], "k:")
                plt.plot(features1[x], features1.status, 'o-', label = 'Train')
                
                plt.plot(featuresT1[x], featuresT1.status,'o-', label = 'Test')
                plt.xlabel('Predicted P(app)')
                plt.ylabel('Actual P(app)')
                #plt.xlim(0.2, 0.8)
                #plt.ylim(0.2, 0.8)
                plt.legend()
                plt.title('Callibration Curve for P(app)')
        
            if x == 'status':
               
                plt.hist(y_probT, range=[0, 1], bins=20, histtype="step", lw=2, label = 'Train')
                plt.hist(y_prob, range=[0, 1], bins=20, histtype="step", lw=2, label = 'Test')
                #plt.xlim(0.2, 0.8)
                plt.ylabel('Count')
                plt.xlabel('P(app)')
                plt.legend()
                plt.title('Histogram of P(app)')
            
            pdf.savefig(fig)
```
