```python
df = df.sample(frac=1)
```

```python
features = df.loc[:,'Variable1':'Variable9']
target =  df['TargetVariable'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.metrics import brier_score_loss

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.30, random_state=143,stratify=target)
```

```python
from sklearn.ensemble import GradientBoostingClassifier as GB
from sklearn.linear_model import LogisticRegression as LG
from sklearn.linear_model import LinearRegression as Lin
#clf = AdaBoostClassifier(GB(),
#                         algorithm="SAMME",
#                         n_estimators=100)
clf = GB(learning_rate=0.05, n_estimators=68)#GB(n_estimators = 100)
#LG()
#RandomForestClassifier(random_state=0, n_jobs=-1)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()

mat = confusion_matrix(y_test, y_pred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
from sklearn import metrics
print(metrics.classification_report(y_pred, y_test))
```

```python
y_prob = clf.predict_proba(X_test)[:,1]
y_probT = clf.predict_proba(X_train)[:,1]

from sklearn.calibration import calibration_curve
fraction_of_positives, mean_predicted_value = \
calibration_curve(y_test, y_prob, n_bins=10)

fraction_of_positivesT, mean_predicted_valueT = \
calibration_curve(y_train, y_probT, n_bins=10)

plt.figure(figsize=(9, 9))
ax1 = plt.subplot2grid((3, 1), (0, 0), rowspan=2)
ax2 = plt.subplot2grid((3, 1), (2, 0))

ax1.plot([0, 1], [0, 1], "k:", label="Perfectly calibrated")
ax1.plot(mean_predicted_value, fraction_of_positives, "s-")
ax1.plot(mean_predicted_valueT, fraction_of_positivesT)

ax2.hist(y_prob, range=(0, 1), bins=50, histtype="step", lw=2)
ax2.hist(y_probT, range=(0, 1), bins=50, histtype="step", lw=2)
```

```python
print(brier_score_loss(y_test, y_prob, sample_weight=None, pos_label=None))
print(brier_score_loss(y_train, y_probT, sample_weight=None, pos_label=None))
```

```python
from matplotlib.backends.backend_pdf import PdfPages

X_train['status'] = np.transpose(y_train)
X_train['pred'] = np.transpose(y_probT)

X_test['status'] = np.transpose(y_test)
X_test['pred'] = np.transpose(y_prob)

with PdfPages('TeleGBM.pdf') as pdf:
    for x in X_train.columns:
        if x != 'test':
        
            X_train['test'] = pd.cut(X_train[x].rank(method='average'),10, labels=False)
            X_train1 = X_train.groupby(['test'], as_index=False).mean()
            X_test['test'] = pd.cut(X_test[x].rank(method='average'),10, labels=False)
            X_test1 = X_test.groupby(['test'], as_index=False).mean()
            fig = plt.figure(x)
            
            if x != 'status' and x != 'pred':
                
                plt.scatter(X_train1[x], X_train1.status, label = 'Train')
                plt.plot(X_train1[x], X_train1['pred'], label = 'Train_Prediction')
                plt.scatter(X_test1[x], X_test1.status, label = 'Test')
                plt.plot(X_test1[x], X_test1['pred'], label = 'Test_Prediction')
                plt.xlabel(x)
                plt.ylabel('P(app)')
                plt.legend()
                plt.title('P(app) as function of '+str(x))
                
            if x == 'pred':
        
                plt.plot([0, 1], [0, 1], "k:")
                plt.plot(X_train1[x], X_train1.status, 'o-', label = 'Train')
                
                plt.plot(X_test1[x], X_test1.status,'o-', label = 'Test')
                plt.xlabel('Predicted P(app)')
                plt.ylabel('Actual P(app)')
                #plt.xlim(0.2, 0.8)
                #plt.ylim(0.2, 0.8)
                plt.legend()
                plt.title('Callibration Curve for P(app)')
        
            if x == 'status':
               
                plt.hist(y_probT, range=[0, 1], bins=20, histtype="step", lw=2, label = 'Train')
                plt.hist(y_prob, range=[0, 1], bins=20, histtype="step", lw=2, label = 'Test')
                #plt.xlim(0.2, 0.8)
                plt.ylabel('Count')
                plt.xlabel('P(app)')
                plt.legend()
                plt.title('Histogram of P(app)')
            
            pdf.savefig(fig)
```

```python
from sklearn import metrics

fpr, tpr, _ = metrics.roc_curve(y_test, y_prob)

plt.plot(fpr, tpr)
plt.plot([0,1],[0,1],'g-')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()
auc = metrics.auc(fpr,tpr)
print('auc = ', auc)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import permutation_test_score

cv = StratifiedKFold(10)

score, permutation_scores, pvalue = permutation_test_score(clf, features.values, target, cv=cv, scoring="roc_auc", n_permutations=100, n_jobs=1)

print("Classification score %s (pvalue : %s)" % (score, pvalue))

plt.hist(permutation_scores, 20, label='Permutation scores',
         edgecolor='black')
ylim = plt.ylim()
plt.plot(2 * [score], ylim, '--g', linewidth=3,
         label='Classification Score'
         ' (pvalue %s)' % pvalue)
plt.plot(2 * [1. / 2], ylim, '--k', linewidth=3, label='Luck') #2 = number of classes

plt.ylim(ylim)
plt.legend()
plt.xlabel('Score')
```
